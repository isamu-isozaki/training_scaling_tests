```python
import pandas as pd
from plot import plot_scaling
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
font = {'family' : 'DejaVu',
        'weight' : 'normal',
        'size'   : 13}
matplotlib.rc('font', **font)
```

```python
df = pd.read_csv("muse_scaling.csv")
df['fp'] = df.mixed_precision.apply(lambda m:'FP16' if m=='--use-mixed-precision' else 'FP32')
#df = df[df.gpus_per_node==4]
df['gpus'] = df['nodes']*df['gpus_per_node']
mask = (df.gpus_per_node==1)&(df.nodes>1)
df = df[~mask]
```

```python
df.head()
```

```python
efficiency_params = {
    "ymin": 800,
    "ymax": 1000,
    "normalize": False,
}
_ = plot_scaling(
    df, 
    name_col="fp",
    node_col="gpus", 
    plot_scaling=True, 
    plot_efficiency=True, 
    plot_ideal=True,
    plot_raw=False, 
    imagespersec_col="imagespersec", 
    figsize=(12,10), 
    efficiency_params=efficiency_params,
)

plt.xlabel("Number of GPUs")
plt.ylabel("Speedup")
plt.show()
```

```python
_ = plot_scaling(
    df, 
    name_col="fp",
    node_col="gpus", 
    plot_scaling=True, 
    plot_efficiency=False, 
    plot_ideal=True,
    plot_raw=False, 
    imagespersec_col="imagespersec", 
    figsize=(16,10), 
)
plt.xlabel("Number of GPUs")
plt.ylabel("Speedup")
plt.show()
```

```python
_ = plot_scaling(
    df, 
    name_col="fp",
    node_col="gpus", 
    plot_scaling=False, 
    plot_efficiency=True, 
    plot_ideal=False,
    plot_raw=False, 
    imagespersec_col="imagespersec", 
    figsize=(12,10), 
    efficiency_params={
        "normalize": False,
    }
)
plt.xlabel("Number of GPUs")
plt.ylabel("Efficiency")
plt.show()
```

```python
baseline_fp32 = df[(df.gpus==1)&(df.fp=='FP32')].imagespersec.values[0] # 1 gpu 1 node throughput for fp32
baseline_fp16 = df[(df.gpus==1)&(df.fp=='FP16')].imagespersec.values[0] # 1 gpu 1 node throughput for fp16

A = df.copy()
a = A[A.fp=="FP32"].copy()
a["fp"] = "FP32(ideal)"
a["imagespersec"] = baseline_fp32 * a["gpus"]

b = A[A.fp=="FP16"].copy()
b["fp"] = "FP16(ideal)"
b["imagespersec"] = baseline_fp16 * b["gpus"]

c = pd.DataFrame([{"gpus":1, "fp": "FP32", "imagespersec": baseline_fp32},{"gpus":1, "fp": "FP32(ideal)", "imagespersec": baseline_fp32}])
d = pd.DataFrame([{"gpus":1, "fp": "FP16", "imagespersec": baseline_fp16}, {"gpus":1, "fp": "FP16(ideal)", "imagespersec": baseline_fp16}])

e = A.copy()
A = pd.concat((a,b,c,d,e))
fig = plt.figure(figsize=(10,8))
p = ["#ff0000", "#F08080", "#3cb44b", "#aaffc3"]
sns.barplot(x="gpus", y="imagespersec", hue="fp",data=A, hue_order=["FP16","FP16(ideal)", "FP32", "FP32(ideal)"], palette=p)
plt.xlabel("Number of GPUs")
plt.ylabel("Throughput (Im/s)")
plt.legend(title="Training mode")
```
